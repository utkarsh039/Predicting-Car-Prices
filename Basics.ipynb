# Introduction to dataset

```python
import pandas as pd
import numpy as np
#to view more columns at a time
pd.options.display.max_columns=99
```

```python
cols = ['symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 
        'drive-wheels', 'engine-location', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-type', 
        'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
cars=pd.read_csv('imports-85.data',names=cols)
```

```python
cars
```

```python
numeric_values_cols = ['normalized-losses', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
numeric_cars=cars[numeric_values_cols]
```

```python
numeric_cars.head(5)
```

# Data Cleaning

replace '?' with 'nan'

```python
numeric_cars=numeric_cars.replace('?',np.nan)
numeric_cars
```

```python
numeric_cars=numeric_cars.astype('float')
numeric_cars.isnull().sum()
```

```python
numeric_cars=numeric_cars.dropna(subset=['price'])
numeric_cars=numeric_cars.fillna(numeric_cars.mean())
numeric_cars.isnull().sum()
```

```python
price_cols=numeric_cars['price']
numeric_cars=(numeric_cars-numeric_cars.min())/(numeric_cars.max()-numeric_cars.min())
numeric_cars['price']=price_cols
numeric_cars.head(5)
```

# Univariate k-nearest neighbors models

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

def knn_train_test(train_col,target_col,df):
    knn=KNeighborsRegressor()
    np.random.seed(1)
    
    #randomize the order of rows 
    shuffled_index=np.random.permutation(df.index)
    random_df=df.reindex(shuffled_index)
    
    # defining train and test indexex
    last_train_row=int(len(random_df)/2)
    train_df=random_df.iloc[0:last_train_row]
    test_df=random_df.iloc[last_train_row:]
    
    #fit he model
    knn.fit(train_df[[train_col]],train_df[target_col])
    
    #prediction using model
    predicted_labels=knn.predict(test_df[[train_col]])
    #Calculation of accuracy or error
    mse=mean_squared_error(test_df[target_col],predicted_labels)
    rmse=mse**(1/2)
    return rmse

rmse_results={}
train_cols=numeric_cars.columns.drop('price')

#now calculate result through each training feature to predict price
for col in train_cols:
    value=knn_train_test(col,'price',numeric_cars)
    rmse_results[col]=value
    
#Make a series object of dictionary 
rmse_results_series=pd.Series(rmse_results)

#sort to get the optimal feature
rmse_results_series.sort_values()
```

# Finding the optimal nearest neighbors

```python
def knn_train_test(train_col,target_col,df):
    np.random.seed(1)
    
    #randomize the order of rows 
    shuffled_index=np.random.permutation(df.index)
    random_df=df.reindex(shuffled_index)
    
    # defining train and test indexex
    last_train_row=int(len(random_df)/2)
    train_df=random_df.iloc[0:last_train_row]
    test_df=random_df.iloc[last_train_row:]
    
    k_values=[1,3,5,7,9]
    k_rmses={}
    
    for k in k_values:
        #fit he model
        knn=KNeighborsRegressor(n_neighbors=k)
        knn.fit(train_df[[train_col]],train_df[target_col])
    
        #prediction using model
        predicted_labels=knn.predict(test_df[[train_col]])
        #Calculation of accuracy or error
        mse=mean_squared_error(test_df[target_col],predicted_labels)
        rmse=mse**(1/2)
        k_rmses[k]=rmse
    return k_rmses


k_rmse_results={}
train_cols=numeric_cars.columns.drop('price')

#now calculate result through each training feature to predict price
for col in train_cols:
    value=knn_train_test(col,'price',numeric_cars)
    k_rmse_results[col]=value

k_rmse_results    
```

# Visualizing through scatter plots

```python
import matplotlib.pyplot as plt
%matplotlib inline

# plot graph on values of dictionary which are also the dictionaries having key as 'k' and values as 'rmse' values
for index,value in k_rmse_results.items():
    x=list(value.keys())
    y=list(value.values())
    
    plt.plot(x,y)
    plt.xlabel('k_values')
    plt.ylabel('RMSE')
```

```python
# Compute average RMSE across different `k` values for each feature.
feature_avg_rmse = {}
for index,value in k_rmse_results.items():
    avg_rmse = np.mean(list(value.values()))
    feature_avg_rmse[index] = avg_rmse
series_avg_rmse = pd.Series(feature_avg_rmse)
series_avg_rmse.sort_values()
```

# Multivariate Model

```python
def knn_train_test(train_col,target_col,df):
    np.random.seed(1)
    
    #randomize the order of rows 
    shuffled_index=np.random.permutation(df.index)
    random_df=df.reindex(shuffled_index)
    
    # defining train and test indexex
    last_train_row=int(len(random_df)/2)
    train_df=random_df.iloc[0:last_train_row]
    test_df=random_df.iloc[last_train_row:]
    
    k=5
    k_rmses={}
    
    
    #fit he model
    knn=KNeighborsRegressor(n_neighbors=k)
    knn.fit(train_df[train_col],train_df[target_col])
    
    #prediction using model
    predicted_labels=knn.predict(test_df[train_col])
    #Calculation of accuracy or error
    mse=mean_squared_error(test_df[target_col],predicted_labels)
    rmse=mse**(1/2)
    k_rmses[k]=rmse
    return k_rmses


k_rmse_results={}


two_best_features = ['horsepower', 'width']
rmse_value = knn_train_test(two_best_features, 'price', numeric_cars)
k_rmse_results['two best features'] = rmse_value

three_best_features = ['horsepower', 'width', 'curb-weight']
rmse_value = knn_train_test(three_best_features, 'price', numeric_cars)
k_rmse_results['three best features'] = rmse_value

four_best_features = ['horsepower', 'width', 'curb-weight', 'city-mpg']
rmse_value = knn_train_test(four_best_features, 'price', numeric_cars)
k_rmse_results['four best features'] = rmse_value

five_best_features = ['horsepower', 'width', 'curb-weight' , 'city-mpg' , 'highway-mpg']
rmse_value = knn_train_test(five_best_features, 'price', numeric_cars)
k_rmse_results['five best features'] = rmse_value

six_best_features = ['horsepower', 'width', 'curb-weight' , 'city-mpg' , 'highway-mpg', 'length']
rmse_value = knn_train_test(six_best_features, 'price', numeric_cars)
k_rmse_results['six best features'] = rmse_value

k_rmse_results
```

# Variation of hyperparameter for top 3 models

```python
def knn_train_test(train_cols, target_col, df):
    np.random.seed(1)
    
    # Randomize order of rows in data frame.
    shuffled_index = np.random.permutation(df.index)
    rand_df = df.reindex(shuffled_index)

    # Divide number of rows in half and round.
    last_train_row = int(len(rand_df) / 2)
    
    # Select the first half and set as training set.
    # Select the second half and set as test set.
    train_df = rand_df.iloc[0:last_train_row]
    test_df = rand_df.iloc[last_train_row:]
    
    k_values = [i for i in range(1,26)]
    k_rmses = {}
    
    for k in k_values:
        # Fit model using k nearest neighbors.
        knn = KNeighborsRegressor(n_neighbors=k)
        knn.fit(train_df[train_cols], train_df[target_col])

        # Make predictions using model.
        predicted_labels = knn.predict(test_df[train_cols])

        # Calculate and return RMSE.
        mse = mean_squared_error(test_df[target_col], predicted_labels)
        rmse = np.sqrt(mse)
        
        k_rmses[k] = rmse
    return k_rmses

k_rmse_results = {}

three_best_features = ['horsepower', 'width', 'curb-weight']
rmse_value= knn_train_test(three_best_features, 'price', numeric_cars)
k_rmse_results['three best features'] = rmse_value

four_best_features = ['horsepower', 'width', 'curb-weight', 'city-mpg']
rmse_value= knn_train_test(four_best_features, 'price', numeric_cars)
k_rmse_results['four best features'] = rmse_value

five_best_features = ['horsepower', 'width', 'curb-weight' , 'city-mpg' , 'highway-mpg']
rmse_value= knn_train_test(five_best_features, 'price', numeric_cars)
k_rmse_results['five best features'] = rmse_value

k_rmse_results
```
